1. *await*: Pausa la ejecución hasta que una promesa se cumpla en funciones asíncronas.
   Ejemplo: await asyncio.sleep(1)

2. *async*: Define una función asíncrona.
   Ejemplo: `async def funcion_asincrona():
                 await alguna_operacion()`

3. *bytes*: Tipo de dato que representa secuencias de bytes.
   Ejemplo: datos = bytes([50, 100, 150])

4. *callable*: Verifica si un objeto es invocable.
   Ejemplo: `if callable(funcion):
                 funcion()`

5. *chr*: Devuelve el carácter correspondiente a un código ASCII.
   Ejemplo: letra = chr(65)  # 'A'

6. *compile*: Compila código en un objeto de código ejecutable.
   Ejemplo: codigo = compile('print("Hola")', '', 'exec')

7. *complex*: Crea un número complejo.
   Ejemplo: numero = complex(2, 3)  # 2 + 3j

8. *dict*: Crea un diccionario.
   Ejemplo: mi_diccionario = dict(a=1, b=2)

9. *dir*: Muestra los atributos de un objeto.
   Ejemplo: print(dir(objeto))

10. *divmod*: Devuelve el cociente y el resto de una división.
    Ejemplo: resultado = divmod(10, 3)  # (3, 1)

11. *enumerate*: Añade un índice a un iterable.
    Ejemplo: `for indice, valor in enumerate(lista):
                 print(indice, valor)`

12. *eval*: Evalúa una expresión como código Python.
    Ejemplo: resultado = eval('2 + 2')  # 4

13. *exec*: Ejecuta código dinámicamente.
    Ejemplo: exec('print("Hola Mundo")')

14. *filter*: Filtra elementos de un iterable según una función.
    Ejemplo: pares = filter(lambda x: x % 2 == 0, lista)

15. *float*: Convierte un valor a tipo flotante.
    Ejemplo: numero = float("3.14")

16. *format*: Formatea una cadena.
    Ejemplo: print("Hola, {}!".format(nombre))

17. *frozenset*: Crea un conjunto inmutable.
    Ejemplo: conjunto_inmutable = frozenset([1, 2, 3])

18. *getattr*: Obtiene un atributo de un objeto.
    Ejemplo: valor = getattr(objeto, 'atributo')

19. *globals*: Devuelve el diccionario de variables globales.
    Ejemplo: variables_globales = globals()

20. *hasattr*: Comprueba si un objeto tiene un atributo específico.
    Ejemplo: `if hasattr(objeto, 'atributo'):
                 print("Tiene el atributo")`

21. *hash*: Devuelve el valor hash de un objeto.
    Ejemplo: valor_hash = hash("cadena")

22. *help*: Muestra la ayuda sobre un objeto o módulo.
    Ejemplo: help(print)

23. *hex*: Convierte un número a hexadecimal.
    Ejemplo: hexadecimal = hex(255)  # '0xff'

24. *id*: Devuelve el identificador único de un objeto.
    Ejemplo: identificador = id(objeto)

25. *input*: Lee una entrada del usuario.
    Ejemplo: nombre = input("¿Cuál es tu nombre?")

26. *int*: Convierte un valor a entero.
    Ejemplo: numero = int("10")

27. *isinstance*: Comprueba si un objeto es de un tipo determinado.

                 print("Es un entero")`

28. *issubclass*: Comprueba si una clase es una subclase de otra.
    Ejemplo: `if issubclass(Hijo, Padre):
                 print("Es subclase")`

29. *iter*: Devuelve un iterador de un objeto.
    Ejemplo: iterador = iter(lista)
