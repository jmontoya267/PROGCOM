```python
class Pikachu(Pokemon):

  __tipo = "Electrico"

  def __init__(self,nombre,nivel,salud,voltaje_max,amperaje_max,color):
    super().__init__(nombre=nombre,nivel=nivel,salud=salud,color=color)
    self.__voltaje_max = voltaje_max
    self.__amperaje_max = amperaje_max


  def salud(self):
    return self.__Pokemon__salud

  def salud(self,salud):
    if salud > 0 and salud <= 100:
      self.__Pokemon__salud = salud
    else:
      print("La salud no puede ser ser negatuva ó la salud no puede ser mayor a 100")

  def nivel(self):
    return self.__Pokemon__nivel

  def nivel(self,nivel):
    if nivel > 0 and nivel <= 100:
      self.__Pokemon__nivel = nivel
    else:
      print("El nivel no puede ser ser negatuvo ó el nivel no puede ser mayor a 100")

  def nombre(self):
    return self.__Pokemon__nombre

  def color(self):
    return self.__Pokemon__color


```

```python

class Humano:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        self.complexion = "Débil"
        self.poder = 150
        self.tipo_arma = "Armas y artes marciales"

    def atacar(self):

        return f"{self.nombre} lanza un ataque básico usando {self.tipo_arma}."

    def mostrar_info(self):
        print(f"{self.nombre} es un humano de {self.edad} años, complexión {self.complexion} y poder de lucha: {self.poder}.")


class Sayajin:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        self.complexion = "Fuerte"
        self.poder = 5000
        self.cola = True
        self.color_pelo = "Negro"
        self.color_ojos = "Negros"

    def atacar(self):

        return f"{self.nombre} lanza un Kamehameha con un poder de {self.poder}!"

    def transformarse(self):

        self.poder *= 2
        return f"{self.nombre} se ha transformado, su poder ahora es {self.poder}."

    def mostrar_info(self):
        print(f"{self.nombre} es un Sayajin de {self.edad} años, complexión {self.complexion}, cabello {self.color_pelo} y ojos {self.color_ojos}. Poder de lucha: {self.poder}.")


class Hibrido(Humano, Sayajin):
    def __init__(self, nombre, edad, madre_raza):

        Humano.__init__(self, nombre, edad)
        Sayajin.__init__(self, nombre, edad)

        self.raza_padre = "Sayajin"
        self.raza_madre = madre_raza


        self.poder = self.poder * 0.6
        self.poder_hibrido = self.poder

    def atacar(self):

        if self.raza_madre == "Humano":
            return f"{self.nombre} lanza un Kamehameha con un poder de {self.poder_hibrido}!"
        else:
            return f"{self.nombre} ataca con su espada Z con un poder de {self.poder_hibrido}!"

    def mostrar_info(self):

        print(f"{self.nombre} es un híbrido de {self.raza_madre} y {self.raza_padre}, de {self.edad} años.")
        print(f"Complexión: {self.complexion}. Poder de lucha: {self.poder_hibrido}.")
        if self.raza_madre == "Humano":
            print(f"{self.nombre} tiene poderes especiales de Sayajin, como el Kamehameha.")
        else:
            print(f"{self.nombre} tiene habilidades de lucha como su padre, incluyendo el uso de la espada Z.")




goku = Sayajin("Goku", 40)
vegeta = Sayajin("Vegeta", 40)


gohan = Hibrido("Gohan", 24, "Humano")
trunks = Hibrido("Trunks", 20, "Humano")


print("\nInformación de Goku (Sayajin):")
goku.mostrar_info()
print(goku.atacar())

print("\nInformación de Vegeta (Sayajin):")
vegeta.mostrar_info()
print(vegeta.atacar())
print(vegeta.transformarse())

print("\nInformación de Gohan (Híbrido):")
gohan.mostrar_info()
print(gohan.atacar())

print("\nInformación de Trunks (Híbrido):")
trunks.mostrar_info()
print(trunks.atacar())
```

```python

class Simpson:
    def __init__(self, nombre):
        self.nombre = nombre
        self.color_cabello = "Rubio"
        self.inteligencia = 50

    def hablar(self):
        return f"{self.nombre} dice: ¡D'oh!"

    def hacer_travesura(self):
        return f"{self.nombre} está haciendo una travesura, ¡a correr!"


class Bouvier:
    def __init__(self, nombre):
        self.nombre = nombre
        self.inteligencia = 100


    def hablar(self):
        return f"{self.nombre} dice: '¡Hola, querido!'"

    def inteligente(self):
        return f"{self.nombre} resuelve un problema de matemáticas complejas."

    def trabajar(self):
        return f"{self.nombre} está trabajando como ejecutiva."


class Liza(Simpson, Bouvier):
    def __init__(self, nombre):
        Simpson.__init__(self, nombre)
        Bouvier.__init__(self, nombre)
        self.inteligencia = Bouvier(self.nombre).inteligencia


    def hablar(self):
        return f"{self.nombre} dice: 'Puedo resolver ecuaciones diferenciales, papá.'"


    def hacer_travesura(self):
        return f"{self.nombre} se está preparando para un experimento científico."


class Bart(Simpson, Bouvier):
    def __init__(self, nombre):
        Simpson.__init__(self, nombre)
        Bouvier.__init__(self, nombre)
        self.inteligencia = Simpson(self.nombre).inteligencia


    def hablar(self):
        return f"{self.nombre} dice: '¡No quiero hacer la tarea!'"


    def hacer_travesura(self):
        return f"{self.nombre} está haciendo una gran travesura: ¡trollear al director Skinner!"


class Maggie(Simpson, Bouvier):
    def __init__(self, nombre):
        Simpson.__init__(self, nombre)
        Bouvier.__init__(self, nombre)
        self.inteligencia = Bouvier(self.nombre).inteligencia * 0.8
    def hablar(self):
        return f"{self.nombre} está balbuceando, pero se nota su inteligencia precoz."

    def hacer_travesura(self):
        return f"{self.nombre} está jugando con sus juguetes y causando algo de caos."




homer = Simpson("Homer Simpson")
marge = Bouvier("Marge Bouvier")


liza = Liza("Liza Simpson")
bart = Bart("Bart Simpson")
maggie = Maggie("Maggie Simpson")


print("\n--- Personajes de la familia Simpson ---")
print(homer.hablar())
print(marge.hablar())

print("\n--- Personajes Híbridos ---")
print(liza.hablar())
print(liza.hacer_travesura())
print(bart.hablar())
print(bart.hacer_travesura())
print(maggie.hablar())
print(maggie.hacer_travesura())
```
